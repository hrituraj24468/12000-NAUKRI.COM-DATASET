# Univariate


# job role

```python
role=data['Job_Role']
role2=role
role=role.unique()
types=[]
typedat=[]
count=0
for datas in role:
    count=0
    for things in role2:
        if things==datas:
            count=count+1
    types.append(datas)
    typedat.append(count)
print(types)
print(typedat)
```

```python
sns.barplot(x=types, y=typedat)
plt.show()
```

```python
rangetype=[]
rangetypedat=[]
i=0
for datasrn in typedat:
    if datasrn>40:
        rangetypedat.append(datasrn)
        rangetype.append(types[i])
    i=i+1
print(rangetype)
print(rangetypedat)
```

```python
sns.barplot(x=rangetype, y=rangetypedat)
plt.show()
```

```python
sns.barplot(x=rangetype[0:5], y=rangetypedat[0:5])
plt.show()
```

# skills description

```python
sd=data['Skills/Description']
skilldatas=[]
for datass in sd:
    sdk=(datass.split(','))
    for dtas in sdk:
        skilldatas.append(str(dtas))
        
print(skilldatas)
```

```python
len(skilldatas)
```

```python
res = [*set(skilldatas)]
print(res)
```

```python
len(res)
```

```python
role=res
role2=skilldatas
types=[]
typedat=[]
count=0
for datas in role:
    count=0
    for things in role2:
        if things==datas:
            count=count+1
    types.append(datas)
    typedat.append(count)
print(types)
print(typedat)
```

```python
rangetype=[]
rangetypedat=[]
i=0
for datasrn in typedat:
    if datasrn>40:
        rangetypedat.append(datasrn)
        rangetype.append(types[i])
    i=i+1
print(rangetype)
print(rangetypedat)
```

```python
sns.barplot(x=rangetype, y=rangetypedat)
plt.show()
```
